```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
path_url = 'https://github.com/ekayuniar/datamining/raw/master/UAS/dataset%20soal%20no.3.xlsx'
data = pd.read_excel(path_url)
df = pd.DataFrame(data)

print(data.head())

print(data.columns)
```

```python

X = data[['Usia', ' Kelahiran ke-', 'Waktu Kelahiran', 'Tekanan Darah',
       'Kelainan Jantung']]
y = data[' Cessar']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_scaled, y)

test_point_1 = np.array([[30, 1, 0, 1,0]])
test_point_2 = np.array([[29,2,0,2,0]])

test_point_1_scaled = scaler.transform(test_point_1)
test_point_2_scaled = scaler.transform(test_point_2)

distances_1, indices_1 = knn.kneighbors(test_point_1_scaled, n_neighbors=5)
distances_2, indices_2 = knn.kneighbors(test_point_2_scaled, n_neighbors=5)

nearest_neighbors_1 = df.iloc[indices_1[0]]
nearest_neighbors_2 = df.iloc[indices_2[0]]

print("5 Nearest Neighbors for test point 1:")
print(nearest_neighbors_1)

print("\n5 Nearest Neighbors for test point 2:")
print(nearest_neighbors_2)
```

```python
print("Hasil prediksi untuk semua data:")
print("-" * 30)
for index, row in df.iterrows():
    prediction = knn.predict([row[X.columns]])[0]
    print("Data ke-", index+1)
    print("Usia:", row["Usia"])
    print("Kelahiran ke:", row[" Kelahiran ke-"])
    print("Waktu Kelahiran:", "HPL" if row["Waktu Kelahiran"] == 0 else ("Prematur" if row["Waktu Kelahiran"] == 1 else "Lewat HPL"))
    print("Tekanan Darah:", "Normal" if row["Tekanan Darah"] == 1 else ("Rendah" if row["Tekanan Darah"] == 0 else "Tinggi"))
    print("Kelainan Jantung:", "Normal" if row["Kelainan Jantung"] == 0 else "Kelainan")
    print("Persalinan:", "Normal" if row[" Cessar"] == 0 else " Cessar")
    print("-" * 30)
```
